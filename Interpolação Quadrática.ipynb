```python
import numpy as np
import matplotlib.pyplot as plt

# Entrada dos três pontos
print("Digite três pontos para interpolação quadrática:")
x_vals = []
y_vals = []
for i in range(3):
    x = float(input(f"x[{i}]: "))
    y = float(input(f"y[{i}]: "))
    x_vals.append(x)
    y_vals.append(y)

# Montagem do sistema Ax = y, com x = [a, b, c]
A = np.array([[x**2, x, 1] for x in x_vals])
b = np.array(y_vals)

# Resolução do sistema
coef = np.linalg.solve(A, b)
a, b, c = coef

print(f"\nPolinômio encontrado: P(x) = {a:.4f}x² + {b:.4f}x + {c:.4f}")

# Estimar y em um ponto fornecido pelo usuário
x_est = float(input("\nDigite o valor de x para estimar y: "))
y_est = a * x_est**2 + b * x_est + c
print(f"Estimativa: y({x_est}) ≈ {y_est:.4f}")

# Geração de pontos para o gráfico
x_interp = np.linspace(min(x_vals)-1, max(x_vals)+1, 500)
y_interp = a * x_interp**2 + b * x_interp + c

# Plot
plt.figure(figsize=(10, 6))
plt.plot(x_interp, y_interp, label='Polinômio Quadrático', color='blue')
plt.scatter(x_vals, y_vals, color='red', label='Pontos Dados')
plt.scatter(x_est, y_est, color='green', label=f'Estimativa: y({x_est:.2f}) ≈ {y_est:.2f}', zorder=5)
plt.title('Interpolação Polinomial Quadrática (por sistema linear)')
plt.xlabel('x')
plt.ylabel('P(x)')
plt.grid(True)
plt.legend()
plt.show()
```
